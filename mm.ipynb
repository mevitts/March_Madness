Regarding EM vs KADJ T and tournament performance, it seems the later rounds reached do match the "Trapezoid of Excellence," as an upside down trapezoid that seems to range KADJ EM of ~(15, 29) and KADJ T ~(62.5, 72.5), and as KADJ EM rises, the wider the possible range can be. 

Next, the lower KADJ O paired with high KADJ D seem to be the most successful in reaching later rounds. Even teams with extremely large KADJ O and decent D do not reach later rounds. BADJ O vs BADJ D is very similar

BARTHAG vs WAB has an extremely positive correlation, and represents sort of a tangential pattern

Typically, first round winners had higher OREB% and EFG% and lower TOV% than losers. They seem to be more effective than FTR and 3P%.

```python
h2h = h2h.rename(columns={'SCORE_D': 'SCORE_A'})

def calculate_differences(row):
    t1_suffix = '_A' if row['Team1'] == 'A' else '_B'
    t2_suffix = '_B' if row['Team1'] == 'A' else '_A'
    
    diff_O = row[f'KADJ O{t1_suffix}'] - row[f'KADJ O{t2_suffix}']
    diff_D = row[f'KADJ D{t1_suffix}'] - row[f'KADJ D{t2_suffix}']
    
    #diffo_t1 = row[f'KADJ O{t1_suffix}'] - row[f'KADJ D{t2_suffix}']
    #diffo_t2 = row[f'KADJ O{t2_suffix}'] - row[f'KADJ D{t1_suffix}']
    
    diffTO_t1 = row[f'TOV%{t1_suffix}'] - row[f'TOV%D{t2_suffix}']
    diffTO_t2 = row[f'TOV%{t2_suffix}'] - row[f'TOV%D{t1_suffix}']
    
    overall_diff = row[f'BARTHAG{t1_suffix}'] - row[f'BARTHAG{t2_suffix}']
    diff_tempo = row[f'KADJ T{t1_suffix}'] - row[f'KADJ T{t2_suffix}']
    diff_efg = row[f'EFG%{t1_suffix}'] - row[f'EFG%{t2_suffix}']
    #diff_tov = row[f'TOV%{t1_suffix}'] - row[f'TOV%{t2_suffix}']
    diff_oreb = row[f'OREB%{t1_suffix}'] - row[f'OREB%{t2_suffix}']
    diff_3pt = row[f'3PT%{t1_suffix}'] - row[f'3PT%{t2_suffix}']
    diff_exp = row[f'EXP RANK{t1_suffix}'] - row[f'EXP RANK{t2_suffix}']
    team1_bal, team2_bal = row[f'Balance Score{t1_suffix}'], row[f'Balance Score{t2_suffix}']
    team1_elo, team2_elo = row[f'ELO{t1_suffix}'], row[f'ELO{t2_suffix}']
    team1_wab, team2_wab = row[f'WAB{t1_suffix}'], row[f'WAB{t2_suffix}']
    team1_seed, team2_seed = row[f'SEED{t1_suffix}'], row[f'SEED{t2_suffix}']
    
    #this references the 'Trapezoid of Excellence', a metric that has a strong history in predicting MM success
    
    team1_trap = row[f'KADJ T{t1_suffix}'] * row[f'BARTHAG{t1_suffix}']
    team2_trap = row[f'KADJ T{t2_suffix}'] * row[f'BARTHAG{t2_suffix}']
    return pd.Series({
        'Diff_AdjO': diff_O,
        'Diff_AdjD': diff_D,
        #'Diff_Ovr': overall_diff,
        'Diff_Tempo': diff_tempo,
        'Diff_EFG%': diff_efg,
        #'Diff_TOV%': diff_tov,
        'Diff_OREB%': diff_oreb,
        'Diff_Bal': team1_bal - team2_bal,
        'Diff_TOV%_T1': diffTO_t1,
        'Diff_TOV%_T2': diffTO_t2,
        'Diff_ELO': team1_elo - team2_elo,
        #'Diff_WAB': team1_wab - team2_wab,
        'Diff_3PT%': diff_3pt,
        'Diff_Exp': diff_exp,
        'Diff_Trap': team1_trap-team2_trap,
        'Team1_Seed': team1_seed,
        'Team2_Seed': team2_seed
    })
    
    
    '''
    return pd.Series({
        'Diff_Adj_T1': diffo_t1,
        'Diff_Adj_T2': diffo_t2,
        #'Diff_Ovr': overall_diff,
        'Diff_Tempo': diff_tempo,
        'Diff_EFG%': diff_efg,
        'Diff_TOV%': diff_tov,
        'Diff_OREB%': diff_oreb,
        'Diff_Bal': team1_bal - team2_bal,
        'Diff_ELO': team1_elo - team2_elo,
        'Diff_WAB': team1_wab - team2_wab,
        'Diff_3PT%': diff_3pt,
        'Team1_Trap': row[f'KADJ T{t1_suffix}'] * row[f'BARTHAG{t1_suffix}'],
        'Team2_Trap': row[f'KADJ T{t2_suffix}'] * row[f'BARTHAG{t2_suffix}'],
        'Team1_Seed': team1_seed,
        'Team2_Seed': team2_seed
    })
'''

    
h2h = h2h.join(h2h.apply(calculate_differences, axis=1))
#h2h
```

```python
from numpy import NaN


h2h['Is_Giant_Killer'] = 0
for index, row in h2h.iterrows():
    if row['Team1'] == 'A' and row['SEED_B'] < row['SEED_A'] and row['BARTHAG_B'] > 0.7:
        h2h.at[index, 'Is_Giant_Killer'] = 1
    elif row['Team1'] == 'B' and row['SEED_A'] < row['SEED_B'] and row['BARTHAG_A'] > 0.7:
        h2h.at[index, 'Is_Giant_Killer'] = 1

#Find Winner if Target = 1
h2h['Target'] = 0
for index, row in h2h.iterrows():
    if row['YEAR'] < 2025:
        if row['Team1'] == 'A' and row['SCORE_A'] > row['SCORE_B']:
            h2h.at[index, 'Target'] = 1
        elif row['Team1'] == 'B' and row['SCORE_B'] > row['SCORE_A']:
            h2h.at[index, 'Target'] = 1
    else:
        h2h.at[index, 'Target'] = NaN

#h2h
```

Adding Weights to upsets

```python
import xgboost as xgb
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer, log_loss

h2h_past = h2h[h2h['YEAR'] < 2025].copy()
h2h_2025 = h2h[h2h['YEAR'] == 2025].copy()

#adds weights upsets to make more of a possibility
h2h_past['Is_Upset'] = (h2h_past['Target'] == 0) & (h2h_past['Team2_Seed'] > h2h_past['Team1_Seed'] + 2)
h2h_past['Weight'] = h2h_past['Is_Upset'].apply(lambda x: 1.3 if x else 1)
```

```python
h2h_past['Seed_Diff'] = h2h_past['Team1_Seed'] - h2h_past['Team2_Seed']
upset_rates = h2h_past.groupby('Seed_Diff').apply(lambda x: (x['Target'] == 0).mean()).to_dict()
h2h_past['Upset_Rate'] =h2h_past['Seed_Diff'].map(upset_rates)

h2h_2025['Seed_Diff'] = h2h_2025['Team1_Seed'] - h2h_2025['Team2_Seed']
h2h_2025['Upset_Rate'] = h2h_2025['Seed_Diff'].map(upset_rates)
upset_rates
```

```python
model2.fit(X_train, y_train, sample_weight=sample_weight)
```

```python
#for each matchup
def matchup_features(team1, team2, final_feats, upset_rates):
    features = {}
    features['Diff_AdjO'] = team1['KADJ O'] - team2['KADJ O']
    features['Diff_AdjD'] = team1['KADJ D'] - team2['KADJ D']
    features['Diff_TOV%_T1'] = team1['TOV%'] - team2['TOV%D']
    features['Diff_TOV%_T2'] = team2['TOV%'] - team1['TOV%D']
    features['Diff_Exp'] = team1['EXP RANK'] - team2['EXP RANK']
    features['Diff_Tempo'] = team1['KADJ T'] - team2['KADJ T']
    features['Diff_EFG%'] = team1['EFG%'] - team2['EFG%']
    features['Diff_OREB%'] = team1['OREB%'] - team2['OREB%']
    features['Diff_Bal'] = team1['Balance Score'] - team2['Balance Score']
    features['Diff_Trap'] = (team1['KADJ T'] * team1['BARTHAG']) - (team2['KADJ T'] * team2['BARTHAG'])
    features['Team1_Seed'] = team1['SEED']
    features['Team2_Seed'] = team2['SEED']
    features['Team1_juggernaut'] = team1['juggernaut']
    features['Team2_juggernaut'] = team2['juggernaut']
    #features['Team1_ff_cal'] = team1['ff_cal']
    #features['Team2_ff_cal'] = team2['ff_cal']
    #features['Team1_champ_caliber'] = team1['champ_caliber']
    #features['Team2_champ_caliber'] = team2['champ_caliber']
    #features['Diff_WAB'] = team1['WAB'] - team2['WAB']
    features['Diff_ELO'] = team1['ELO'] - team2['ELO']
    
    features['Seed_Diff'] = abs(team1['SEED'] - team2['SEED'])
    features['Upset_Rate']= upset_rates.get(features['Seed_Diff'], 0)
    
    features['Is_Giant_Killer'] = 1 if team1['SEED'] > team2['SEED'] and team2['BARTHAG'] > 0.65 else 0
        
    return pd.Series(features)[final_feats]
    
```

```python
#get play in games
teams = merged_df_processed[merged_df_processed['YEAR'] == 2025].copy()
play_in = teams.groupby(['QUAD NO', 'SEED']).filter(lambda x: len(x) > 1)

play_in_matchups = play_in.merge(play_in, on=['QUAD NO', 'SEED'], suffixes=('_team1', '_team2'))
play_in_matchups = play_in_matchups[play_in_matchups['TEAM_team1'] < play_in_matchups['TEAM_team2']]
#play_in_matchups.columns
```

```python
play_in_results = []
for index, game in play_in_matchups.iterrows():
    team1 = teams[teams['TEAM'] == game['TEAM_team1']].iloc[0]
    team2 = teams[teams['TEAM'] == game['TEAM_team2']].iloc[0]
    
    X = matchup_features(team1, team2, final_feats, upset_rates)
    prob = model1.predict_proba([X])[0,1]
    winner = ''
    if prob >= 0.5:
        winner = team1
    else:
        winner = team2
    play_in_results.append({'QUAD NO': game['QUAD NO'], 'SEED': game['SEED'], 'Winner': winner})
    
#manually changed result as this game has occurred
play_in_results[0]['Winner'] = teams[teams['TEAM'] == 'Alabama St.'].iloc[0]
play_in_results[2]['Winner'] = teams[teams['TEAM'] == 'North Carolina'].iloc[0]
```

```python
round_1_seeds = [(1, 16), (8, 9), (5, 12), (4, 13), (6, 11), (3, 14), (7, 10), (2, 15)]
quads = teams['QUAD NO'].unique()
games = []

#setting up first round matchups
for quad in quads:
    teams_in_region = teams[teams['QUAD NO'] == quad]
    seed_team = {}
    for i in range(1, 17):
        teams_with_seed = teams_in_region[teams_in_region['SEED'] == i]
        if len(teams_with_seed) > 1:
            winner = next(res['Winner'] for res in play_in_results if (res['QUAD NO'] == quad) and (res['SEED'] == i))
            seed_team[i] = winner
        else:
            seed_team[i] = teams_with_seed.iloc[0]
    for seed1, seed2 in round_1_seeds:
        team1, team2 = seed_team[seed1], seed_team[seed2]
        games.append({'Team 1': team1, 'Team 2': team2, 'Winner': None})
        
games[0]['Team 1']

#X = matchup_features(games[0]['Team 1'], games[0]['Team 2'], final_feats, upset_rates)
```

```python
def predict_round(games, model, final_feats, upset_rates):
    for i in range(len(games)):
        X = matchup_features(games[i]['Team 1'], games[i]['Team 2'], final_feats, upset_rates)
        prob = model.predict_proba([X])[0,1]
        if prob >= 0.5:
            games[i]['Winner'] = games[i]['Team 1']
        else:
            games[i]['Winner'] = games[i]['Team 2']
            
    return games
        
def next_round_setup(games):
    next_round = []
    for i in range(0, len(games), 2):
        team1 = games[i]['Winner']
        team2 = games[i + 1]['Winner']
        if team1['BARTHAG'] >= team2['BARTHAG']:
            next_round.append({'Team 1': team1, 'Team 2': team2, 'Winner': None})
        else:
            next_round.append({'Team 1': team2, 'Team 2': team1, 'Winner': None})
        
    return next_round
```

```python
rounds = {
    '64': games,
    '32': [],
    '16': [],
    '8': [],
    '4': [],
    '2': []
}

rounds2 = {
    '64': games,
    '32': [],
    '16': [],
    '8': [],
    '4': [],
    '2': []
}
round_nums = ['64', '32', '16', '8', '4', '2']

for i, round in enumerate(round_nums):
    rounds[round] = predict_round(rounds[round], model1, final_feats, upset_rates)
    if round != '2':
        rounds[round_nums[i + 1]] = next_round_setup(rounds[round])
'''
for i, round in enumerate(round_nums):
    rounds2[round_nums[i]] = predict_round(rounds2[round], model2, final_feats, upset_rates)
    if round != '2':
        rounds2[round_nums[i + 1]] = next_round_setup(rounds2[round])
'''
with open('tournament_results.txt', 'w', encoding="utf-8") as file:
    for round_name, games in rounds.items():
        file.write(f"\n{round_name}:\n")
        for idx, game in enumerate(games):
            winner_name = game['Winner']['TEAM']
            team1_name = game['Team 1']['TEAM']
            team2_name = game['Team 2']['TEAM']
            file.write(f"Game {idx + 1}: {team1_name} vs {team2_name} -> Winner: {winner_name}\n")
            '''
with open('tournament_results2.txt', 'w', encoding="utf-8") as file:
    for round_name, games in rounds2.items():
        file.write(f"\n{round_name}:\n")
        for idx, game in enumerate(games):
            winner_name = game['Winner']['TEAM']
            team1_name = game['Team 1']['TEAM']
            team2_name = game['Team 2']['TEAM']
            file.write(f"Game {idx + 1}: {team1_name} vs {team2_name} -> Winner: {winner_name}\n")
        
'''
```
